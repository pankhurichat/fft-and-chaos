```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math as m
import sympy as sy
```

```python
def f(x):
    if x>0 and x<=4:
        return 0
    elif x>4 and x<=7:
        return 1/9
    elif x>7 and x<=10:
        return 2/9
    elif x>10:
        return 0
```

```python
q=[f(u) for u in range(1,11)]
```

```python
plt.plot(q)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x)")
plt.show()
```

```python
def F(x):
    if x>-10000 and x<=4:
        return 0
    elif x>4 and x<=7:
        return (x-4)/9
    elif x>7 and x<=10:
        return (2*x-11)/9
    elif x>10:
        return 1
```

```python
Q=[F(U) for U in range(1,11)]
```

```python
plt.plot(Q)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x)")
plt.show()
```

```python
plt.plot(q,label='PDF')
plt.plot(Q,label='CDF')
plt.legend()
plt.show()
```

```python
def gauss(x,u,s):
    return [1/(2*m.pi*s**2)**0.5*m.e**(-(x[i]-u)**2/(2*s**2)) for i in range(0,len(x))]
```
