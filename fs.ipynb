```python
import numpy as np
import scipy
from scipy import integrate as inte
import matplotlib.pyplot as plt
from numpy import cos,pi,abs,exp,sin
```

```python
def fun(x): #the fucntion we are writing the fourries series for
    #return abs(np.sin(x))
    return exp(x)
```

```python
x=np.linspace(-3.14,3.14,10000) # the limits for which FS is defined 
```

```python
k=1/10000 #dx for integration via simpson method
```

```python
n=50 # no of polynomials required
```

```python
L=(x[-1]-x[0])/2 # L
```

calculating $a_o$

```python
a_o=1/L * inte.simps(fun(x),x,dx=k) 
```

```python
plt.plot(x,fun(x))
```

Calculating the different values for coefficients and storing them into arrays

```python
ans=np.zeros(n)
for i in range(1,n):
    ans[i-1]= 1/L * inte.simps(fun(x)*cos((i*np.pi*x)/L),x,dx=k)

bns=np.zeros(n)
for i in range(1,n):
    bns[i-1]= 1/L * inte.simps(fun(x)*sin((i*np.pi*x)/L),x,dx=k)
```

```python
s=0
for i in range(1,n):
    s+= ans[i-1]*cos((i*pi*x )/L)

for i in range(1,n):
    s+= bns[i-1]*sin((i*pi*x )/L)
    
s+=a_o/2
```

```python
plt.plot(x,s,"blue")
plt.plot(x,fun(x),"black")
```

```python
ans=np.zeros(n)
for i in range(1,n):
    ans[i]= 1/L * inte.simps(fun(x)*cos((i*np.pi*x)/L),x,dx=k)

bns=np.zeros(n)
for i in range(1,n):
    bns[i]= 1/L * inte.simps(fun(x)*sin((i*np.pi*x)/L),x,dx=k)
```

```python
s=0
for i in range(1,n):
    s+= ans[i]*cos((i*pi*x )/L)

for i in range(1,n):
    s+= bns[i]*sin((i*pi*x )/L)
    
s+=a_o/2
```

```python
plt.plot(x,s,"blue")
plt.plot(x,fun(x),"black")
```
