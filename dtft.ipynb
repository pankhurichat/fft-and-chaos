```python
def sorter(arr,p):
    N=len(arr)
    l=int(N/p)
    temp=np.empty((0,p))
    for i in range(p):
        a=np.zeros(int(l/2))
        b=np.zeros(int(l/2))
        for j in range(int(l/2)):
            a[j]=arr[l*i+2*j]
            b[j]=arr[l*i+2*j+1]
        a=np.concatenate((a,b))
        temp=np.append(temp,a)
    return temp
```

```python
x=[1,2,3,4,5,6,7,8,9]
a=sorter(x,2)
```

```python
def transform(arr,p):
    N=len(arr)
    q=int(N/p)
    s=int(p/2)
    temp=np.empty((0,s))
    w=np.e**((-2*np.pi*1j)/(2*q))
    for i in range(s):
        m=i*q*2
        a=np.zeros(2*q,dtype='complex')
        for k in range(q):
            a[k]=arr[m+k]+(w**k)*arr[m+k+q]
            a[k+q]=arr[m+k]-(w**k)*arr[m+k+q]
        temp=np.append(temp,a)
    return temp
```

```python
transform(a,3)
```
